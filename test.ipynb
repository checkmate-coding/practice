Now that the libraries are installed, let's load a pre-trained hate speech detection model from Hugging Face. We'll use a `pipeline` for ease of use.

Now, let's define a prediction function that will use our loaded `hate_speech_detector` pipeline.

Finally, we'll create the Gradio interface and launch it.

```python
import gradio as gr
from transformers import pipeline

# Load the model globally, once, before defining the prediction function
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
hate_speech_detector = pipeline("sentiment-analysis", model=model_name)

def predict_sentiment(text):
    # The hate_speech_detector is now a global variable, loaded once
    result = hate_speech_detector(text)
    # The result is a list of dictionaries, e.g., [{'label': 'POSITIVE', 'score': 0.9998}]
    # We'll extract the label and score for display.
    label = result[0]['label']
    score = result[0]['score']
    return f"Label: {label}, Confidence: {score:.4f}"

iface = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(lines=5, placeholder="Enter text here to detect sentiment..."),
    outputs="text",
    title="SNS Hate Speech Detection (Sentiment Analysis)",
    description="Enter a sentence to see if it's classified as POSITIVE or NEGATIVE by a pre-trained sentiment analysis model. This can serve as a baseline for hate speech detection."
)

iface.launch(debug=True)
```
